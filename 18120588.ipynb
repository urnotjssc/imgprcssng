```python
#Image Processing
#18120588 
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as image
from PIL import Image
```

```python
#read image: 
def read_image(str):
    img = plt.imread(str)
    plt.imshow(img)
    plt.axis('off')
    return img 
```

```python
img = read_image("sample.jpg")
```

```python
#CÂU 1: CHỈNH ĐỘ SÁNG:
```

```python
def NewPixel(pixel, brightnessFactor):
	result = float(pixel * (1 + brightnessFactor))
	if result < 0:
		return 0
	elif result > 255:
		return 255
	return int(result)
```

```python
def ChangeBrightness(img, brightnessFac):

    #convert to 2D array:
    x,y,z = img.shape
    img_2d = img.reshape(x*y,z)
    pixel = np.zeros((x*y, z))

    # Change Brightness:
    for i in range (img_2d.shape[0]):
        for j in range (img_2d.shape[1]):
            pixel[i,j] = NewPixel(img_2d[i,j],brightnessFac)
    
    img_result = (np.reshape(pixel,(x,y,z)))/255

    plt.imshow(img_result)
    # save the result:
    plt.imsave('ChangeBrightness('+ str(brightnessFac) +').png', img_result)
    return img_result
```

```python
brightnessFac = float(input("Enter brightness amount (-1.0 to 1.0): "))
ChangeBrightness(img, brightnessFac)
```

```python
#CÂU 2: CHỈNH ĐỘ TƯƠNG PHẢN
```

```python
def Trucate(pixel):
    if pixel < 0:
	    return 0
    elif pixel > 255:
	    return 255
    return pixel
```

```python
def AdjustContrast(img, ContrastFac):

    F = 259*(ContrastFac+255)/(255*259-ContrastFac)

    #convert to 2D array:
    x,y,z = img.shape
    img_2d = img.reshape(x*y,z)
    pixel = np.zeros((x*y, z))

    # Change Contrast:
    
    for i in range (img_2d.shape[0]):
        for j in range (img_2d.shape[1]):
            pixel[i,j] = Trucate(F*(img_2d[i,j]-128)+128)
    pixel = pixel / 255

    img_result = np.reshape(pixel,(x,y,z))

    plt.imshow(img_result)
    # save the result:
    plt.imsave('AdjustContrast('+ str(ContrastFac) +').png', img_result)
    return img_result
```

```python
ContrastFac = int(input("Input contrast amount (-255 to 255): "))
AdjustContrast(img,ContrastFac)
```

```python
def Grayscale(img):

    x,y,z = img.shape
    img_result = np.zeros((x,y,z))

    for i in range (x):
        for j in range (y):

            r, g, b = img[i,j,0], img[i,j,1],img[i,j,2]
            avg = int(round(r*0.3 + g*0.59 + b*0.11))
            img_result[i,j] =[avg, avg, avg]

    img_result = img_result/255
    plt.imshow(img_result)

    plt.imsave('Grayscale.png',img_result)
    return img_result
```

```python
Grayscale(img)
```

```python
def verticalFlip(img):
    R, G, B = img[:,:,0], img[:,:,1], img[:,:,2]
    R = R[::-1]
    G = G[::-1]
    B = B[::-1]

    x,y,z = img.shape
    img_result = np.zeros((x,y,z))

    img_result[:,:,0], img_result[:,:,1], img_result[:,:,2] = R,G,B
    img_result = img_result /255

    plt.imshow(img_result)
    # save:
    plt.imsave('VerticalFlip.png', img_result)
    return img_result
```

```python
verticalFlip(img)
```

```python
def horizontallyFlip(img):
    
    x,y,z = img.shape
    img_result = np.zeros((x,y,z))

    for k in range (z):
        for i in range (x):
            len = y-1
            for j in range (int((y+1)/2)):
                    img_result[i,j,k] = img[i,len,k]
                    img_result[i,len,k] = img [i,j,k]
                    len -=1
    img_result = img_result/255
    plt.imshow(img_result)
    # save:
    plt.imsave('HorizontallyFlip.png', img_result)
    return img_result
```

```python
horizontallyFlip(img)
```

```python
def Overlay(img1 , img2, alpha):
    x,y,z = img1.shape
    #dst = src1 * alpha + src2 * beta 
    beta = 1 - alpha
    x,y,z = img1.shape
    mix_pixel = []
    
    img_result =np.zeros((x,y,z))
    
    mix_pixel = img1 *alpha +img2*beta 
    
    for i in range (x):
        for j in range (y):
            r, g, b = mix_pixel[i,j,0], mix_pixel[i,j,1],mix_pixel[i,j,2]
            avg = int(round(r*0.3 + g*0.59 + b*0.11))
            img_result[i,j] =[avg, avg, avg]
    
    img_result = img_result/255
    plt.imshow(img_result)
    image.imsave('Overlay.png',img_result)
    return img_result
```

```python
background = read_image("Background_flag.jpg")
star = read_image("Star_Flag.jpg")
alpha = float(input("Enter alpha amount (from 0 to 1):"))
Overlay(background,star,alpha)
```

```python
def pixel(matrix,i,j):
    total = 0
    for x in range(i-1,i+2):
        for y in range(j-1,j+2):
            total += matrix[x][y]
    return total//9
```

```python
def boxBlur_layer(layer):
    x,y = layer.shape
    result = np.zeros((x,y))
    for i in range(1,x-1):
        for j in range(1,y-1):
            result[i,j] = (pixel(layer,i,j))
    return result
  
```

```python
def boxBlur_image(img):
    R, G, B = img[:,:,0],img[:,:,1],img[:,:,2]
    newR = boxBlur_layer(R)
    newG = boxBlur_layer(G)
    newB = boxBlur_layer(B)
    
    x,y,z = img.shape
    img_result = np.zeros((x,y,z))

    for i in range (x):
        for j in range (y):
            img_result[i,j,0] , img_result[i,j,1] , img_result[i,j,2] = newR[i,j], newG[i,j], newB[i,j]
    img_result =img_result/255
    plt.imshow(img_result)
    # save:
    plt.imsave('Blur.png', img_result)
    return img_result
          
```

```python
boxBlur_image(img)
```
